1. Curtain company - $6.50/m (make the curtains), $45 (installation)<br>
    Stores: cost of material per metre, unmber of metres the user requires<br>
    Outputs: total cost of cloth, cost of making curtains, total cost of curtains (including installation)

```python
cost = 6.50
metres = float(input('This company charges $6.50 per metre to make a set of curtains, and $45 to install them.\nHow many metres of curtain would you like? '))
print(f'The total cost of cloth and making them into curtains will be ${metres*cost:.2f}.')
print(f'The total fee for curtain making and installation will be ${metres*cost+45:.2f}')
```

2. Plumber - $1.50/km travelled, $60/h of labour, + part costs<br>
   Stores: Distance travelled, number of hours worked, cost of parts<br>
   Outputs: Cost of travel, cost of labour, total cost of the bill 

```python
distance_cost = 1.5
labour_cost = 60

distance = float(input('How many kilometres did you travel? '))
hours = float(input('How many hours did you work? '))
parts = float(input('How much did the parts cost? '))

print(f'The cost of travel is ${distance*distance_cost:.2f}.')
print(f'The cost of labour is ${hours*labour_cost:.2f}.')
print(f'The total bill cost is ${distance*distance_cost + hours*labour_cost + parts:.2f}.')
```

3. Landscape Gardening Firm  - $65/tonne (topsoil), $25 (trailer).<br>
    WAP to calculate the amount paid by customers buying any amount of topsoil they input. This input should be saved in a variable<br>
    Outputs: amount of topsoil purchased, cost of the topsoil, the total cost (with suitable labels so that the user understands the bill)

```python
topsoil_cost = 65
trailer_cost = 25

topsoil = float(input('How many tonnes of topsoil do you want? '))
print(f'You purchased {topsoil} tonnes of topsoil.')
print(f'The cost of all the topsoil is ${topsoil*topsoil_cost:.2f}')
print(f'The total fee (with trailer) will be ${topsoil*topsoil_cost+trailer_cost:.2f}')
```
