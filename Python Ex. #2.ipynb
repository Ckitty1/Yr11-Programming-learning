1. Curtain company - $6.50/m (make the curtains), $45 (installation)<br>
    Stores: cost of material per metre, unmber of metres the user requires<br>
    Outputs: total cost of cloth, cost of making curtains, total cost of curtains (including installation)

```python
cost = 6.50
metres = float(input('This company charges $6.50 per metre to make a set of curtains, and $45 to install them.\nHow many metres of curtain would you like? '))
print(f'The total cost of cloth and making them into curtains will be ${metres*cost:.2f}.')
print(f'The total fee for curtain making and installation will be ${metres*cost+45:.2f}')
```
