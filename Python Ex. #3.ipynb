1. Video Shop  - Rents videos for $2 per. However, if a customer rents 10 videos, they are charged $10 for the first 10 and $2 for each one above that. 
Determine inputs, processing and outputs needed, and write the code.

```python
videos =  int(input('How many videos do you want to rent? '))

if videos >= 10:
    print(f'Your total cost will be ${100 + (videos-10)*2:.2f}.')

else:
    print(f'Your total cost will be ${videos*2:.2f}.')
```

2. WAP that will input someone’s age and print out either “You are not old enough to vote” or “You are old enough to vote” depending on whether they are 18 or not. 

```python
age = int(input('How old are you? '))

if age >= 18:
    print('You are old enough to vote.')
else:
    print('You are not old enough to vote.')
```

3. WAP that will ask the user to input a number. The output should tell the user whether the number is larger or smaller than zero, or equal to zero. 

```python
num = float(input('Please input a number: '))

if num > 0:
    print('Your number is larger than 0.')
elif num < 0:
    print('Your number is smaller than 0.')
else:
    print('Your number is 0.')
```

4. The gas company charges 7.45 cents per unit for the first 55 units of gas and 6.85 cents per unit for each extra unit of gas used. WAP that will calculate a customer’s monthly account when the number of units of gas used is entered. The account should be in dollars 

```python
units = float(input('How many units of gas do you want? '))

if units >= 55:
    print(f'Your total cost will be ${(7.45*55 + (units-55)*6.85)/100:.2f}.')
else:
    print(f'Your total cost will be ${(7.45*units)/100:.2f}.')
```

5. WAP that asks for the length of three sides of a triangle and if it is a valid triangle  (2 sides cannot add up must be more than the third side), calculates the perimeter of the triangle. Otherwise it prints out an error message. 

```python
a = float(input('Please input the length of the first side of the triangle: '))
b = float(input('Please input the length of the second side of the triangle: '))
c = float(input('Please input the length of the third side of the triangle: '))

if a+b>c and a+c>b and b+c>a:
    print(f'The triangle is valid. The perimeter is {a+b+c}.')
else:
    print('The sides do not form a valid triangle. Please try again.')
```
