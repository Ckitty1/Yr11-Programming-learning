# Dictionaries

making a dictionary

```python
dict = {'key1':'value1','key2':'value2'}
dict['key1']
```

dictionaries can also hold other things as values like lists


```python
list_dict = {'key1':[0,1,2,3,4],'key2':12345}
list_dict['key1'][0]
```

```python
list = list_dict['key1']
print(list)
list[2]
```


changing a certain value in the dictionary


```python
dict['key1'] = 'valued'
dict
```

adding a single key and value to the dictionary


```python
dict['key3'] = 'value3'
dict
```

addings multiple keys and their values at the same time


```python
dict.update({'key1':'updated1', 'key2':'updated2'})
dict
```

lists all keys in the dictionary

```python
dict.keys()
```

lists the values only in the dictionary

```python
dict.values()
```

lists them both in pairs

```python
dict.items()
```
