3. WAP that will ask the user for a number, save it in a veriable called num and will then print out the following info using that number, with each statement on a different line.

The number is ___<br>
5 times ___ is ___<br>
if you add 3 to ___ the answer is ___<br>
if you divide 20 by ___, the answer is ___<br>

```python
num = float(input('Please input any number: '))

print(f'The number is {num}.')
print(f'5 times {num} is {num * 5}.')
print(f'If you add 3 to {num}, the answer is {num + 3}.')
print(f'If you divide 20 by {num}, the answer is {20 / num}.')
```

4. WAP that will ask for a particular number in inches and will convert that number to cm with an output like: <br>
___ inches is equal to ___cm

```python
inches = float(input('Please enter an amount of inches: '))

if inches == 1:
    print(f'\n{inches} inch is equal to {inches * 2.54}cm.')

else:
    print(f'\n{inches} inches are equal to {inches * 2.54}cm.')
```

6. WAP which will ask for three numbers, save them in three different variables and will then print out the three numbers, the total and the average in the following format:.<br>
The numbers were ___,___ and ___.<br>
The total of these numbers is ___.<br>
The average of these numbers is ___.<br>

```python
num1 = float(input('Please enter your first number: '))
num2 = float(input('Please enter your second number: '))
num3 = float(input('Please enter your third number: '))

print(f'The numbers were {num1}, {num2} and {num3}.')
print(f'The total of these numbers is {num1 + num2 + num3}.')
print(f'The average of these numbers is {(num1 + num2 + num3)/3}.')
```
